```python
import cv2

# Image preparation with OpenCV
image = cv2.imread('/kaggle/input/ocr-receipts-text-detection/images/17.jpg')

# Returns image in grayscale format
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define the contrast and brightness value
contrast = 1. # Contrast control ( 0 to 127)
brightness = 40. # Brightness control (0-100)

weighted = cv2.addWeighted(gray, contrast, gray, 0, brightness)

cv2.imwrite('result.png', weighted)
```

```python
from PIL import Image, ImageOps
import pytesseract

# Load the image
image_path = 'result.png'
image = Image.open(image_path)

# # Rotate the image (e.g., 90 degrees clockwise)
# rotated_image = image.rotate(90, expand=True)


# Resize the image (e.g., width=600, height=1024)
new_size = (800, 600)
resized_image = image.resize(new_size, Image.ANTIALIAS)

# Convert to grayscale
gray_image = ImageOps.grayscale(resized_image)

# Apply thresholding
threshold = 255  # you can adjust this value
thresholded_image = gray_image.point(lambda p: p > threshold and 255)

# Save the processed image if needed (optional)
processed_image_path = 'processed_image.png'
gray_image.save(processed_image_path)
```
